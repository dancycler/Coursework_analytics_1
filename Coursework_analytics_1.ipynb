```python
### YOUR CODE HERE ###
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
import matplotlib.patches as patches
import random

def calculateAge1(b): #функция для расчета возраста по дате рождения
    today = pd.to_datetime('today')
    birthDate = pd.to_datetime(b)
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day)) 
  
    return age 

dat1 = pd.read_csv("Data/courses.csv")
dat2 = pd.read_csv("Data/students.csv")
dat3 = pd.read_csv("Data/course_contents.csv")
dat4 = pd.read_csv("Data/progresses.csv")
dat5 = pd.read_csv("Data/progress_phases.csv")

datst = dat2.copy(deep=True)
datst = datst.dropna()
datst['bdate'] = pd.to_datetime(datst['birthday'], errors='coerce') #дата рождения
datst = datst.dropna()
datst['age'] = datst['bdate'].apply(calculateAge1) #возраст
datst = datst[datst['age'] > 0] # удаляем строки, в которых возраст является отрицательным

main = pd.merge(dat5, dat4, how ='left', left_on=('progress_id'), right_on=('id'))
del main['id']
main = pd.merge(main, dat1, how ='left', left_on=('course_id'), right_on=('id'))
del main['id']
del main['Unnamed: 0']
main = pd.merge(main, dat3, how = 'left', on = ('course_id', 'module_number', 'lesson_number'))

#выделим отдельный датафрейм (он потребуется в следующих заданиях), в котором будут храниться только ДЗ
main_hw = main[main['is_homework'] == True] 

#Конвертируем дату начала и конца выполнения задания в фомат даты/времени
main_hw['sdate'] = pd.to_datetime(main_hw['start_date'], errors='coerce') 
main_hw['fdate'] = pd.to_datetime(main_hw['finish_date'], errors='coerce')
main_hw = main_hw.dropna(subset=['sdate'])
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

Такой показатель времени можно получить двумя путями: нахождением среднего или нахождением медианы. В нашем случае медиана лучше подходит, оно позволяет игнорировать как ошибочные данные (время выполнения пеньше секунды) так и данные создающие сильный перекос, кроме того не стоит забывать, что не все студенты закончили работы и их тоже нужно оценивать, поэтому будем считать датой окончания их работы последнюю дату в датафрейме (подразумевая, что это дата формирования таблицы). <br>
Найдя медианное время по всем дз их необходимо сравнить с данными студентов. Предлагается просто разделить время выполнения студентом дз на медианное время выполнения того же дз. После этого находится среднее значение этого показателя успеваемости для студента по конкретному курсу (возможен случай, при котором студент учится сразу на нескольких курсах и его успеваемость отличается на них).<br>
Далее полученные показатели успеваемости стоит сравнить с неким числом, которое будет выполнять роль допустимых границ успеваемости. Эмпирически было подобрано число 10. В данном случае студенты все еще успешно справляются с домашними заданиями, хоть и тратят довольно много времени.
